```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Create the dataset
data = {'Rooms': [1, 2, 3, 4, 5, 6],
        'Price': [100000, 150000, 200000, 250000, 300000, 350000]}
df = pd.DataFrame(data)

# 2. Setup and Train the AI
X = df[['Rooms']] 
y = df['Price']
model = LinearRegression()
model.fit(X, y)

# 3. Predict price for a 7-room house (fixed to remove the warning)
new_house = pd.DataFrame([[7]], columns=['Rooms'])
prediction = model.predict(new_house)
print(f"The AI predicts a 7-room house costs: ${prediction[0]:,.2f}")

# 4. Visualizing the "AI Logic"
plt.scatter(df.Rooms, df.Price, color='blue', label='Actual Data')
plt.plot(df.Rooms, model.predict(X), color='red', label='AI Trend Line')
plt.title('Housing Market AI Model')
plt.xlabel('Number of Rooms')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
